```python
actual = np.array([4200, 4300, 3900, 4100, 4500])
predicted = np.array([4000, 4200, 3900, 4100, 4300])

error = actual - predicted
absolute_error = np.abs(error)
squared_error = error ** 2

mae = np.mean(absolute_error)
mse = np.mean(squared_error)
rmse = np.sqrt(mse)
mape = np.mean(np.abs(error / actual)) * 100

ss_res = np.sum((actual - predicted) ** 2)
ss_total = np.sum((actual - np.mean(actual)) ** 2)
r2 = 1 - (ss_res / ss_total)

print("Actual:", actual)
print("Predicted:", predicted)
print("Error:", error)
print("Absolute Error:", absolute_error)
print("Squared Error:", squared_error)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", np.round(rmse, 2))
print("MAPE:", np.round(mape, 2), "%")
print("R²:", np.round(r2, 3))
```

```python
import numpy as np

# 1. Create dataset
data = np.array([
    [10, 30, 45, 4200],
    [12, 35, 50, 4300],
    [9, 40, 42, 3900],
    [11, 32, 47, 4100],
    [13, 29, 52, 4500],
    [14, 28, 55, 4600],
    [8, 42, 40, 3700],
    [15, 27, 56, 4700],
    [10, 36, 44, 4050],
    [12, 33, 49, 4350]
], dtype=float)

print("Original data shape:", data.shape)

# 2. Shuffle data
rng = np.random.default_rng(42)

indices = rng.permutation(len(data))
shuffled_data = data[indices]

# 3. Train-test split
train_size = int(0.8 * len(shuffled_data))

train_data = shuffled_data[:train_size]
test_data = shuffled_data[train_size:]

# 4. Separate X and y
X_train = train_data[:, 0:3]
y_train = train_data[:, 3]

X_test = test_data[:, 0:3]
y_test = test_data[:, 3]

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# 5. Feature scaling
train_mean = np.mean(X_train, axis=0)
train_std = np.std(X_train, axis=0)

X_train_scaled = (X_train - train_mean) / train_std
X_test_scaled = (X_test - train_mean) / train_std

print("Training feature mean:", np.round(train_mean, 2))
print("Training feature std:", np.round(train_std, 2))

print("Scaled X_train:")
print(np.round(X_train_scaled, 2))

print("Scaled X_test:")
print(np.round(X_test_scaled, 2))

# 6. Manual prediction model
weights = np.array([100, -5, 20])
intercept = 2500

y_pred = intercept + np.dot(X_test, weights)

# 7. Error calculation
error = y_test - y_pred
absolute_error = np.abs(error)
squared_error = error ** 2

# 8. Metrics
mae = np.mean(absolute_error)
mse = np.mean(squared_error)
rmse = np.sqrt(mse)
# 9. Results
print("Actual y_test:", y_test)
print("Predicted y_pred:", y_pred)

print("Error:", error)
print("Absolute Error:", absolute_error)
print("Squared Error:", squared_error)

print("MAE:", np.round(mae, 2))
print("MSE:", np.round(mse, 2))
print("RMSE:", np.round(rmse, 2))

if mae <= 100:
    print("Good model")
elif mae <= 200:
    print("Acceptable model")
else:
    print("Model needs improvement")
```

```python
import numpy as np

# 1. Create dataset
data = np.array([
    [10, 30, 45, 4200],
    [12, 35, 50, 4300],
    [9, 40, 42, 3900],
    [11, 32, 47, 4100],
    [13, 29, 52, 4500],
    [14, 28, 55, 4600],
    [8, 42, 40, 3700],
    [15, 27, 56, 4700],
    [10, 36, 44, 4050],
    [12, 33, 49, 4350],
    [np.nan, 34, 48, 4250],
    [16, 26, 57, 4800],
    [7, 45, 39, 3600],
    [13, np.nan, 51, 4450],
    [14, 31, 54, 4550]
], dtype=float)

print("Original data:")
print(data)

print("Original data shape:", data.shape)
print("Number of missing values:", np.sum(np.isnan(data)))

# 2. Clean missing values
column_means = np.nanmean(data, axis=0)

data_clean = data.copy()

missing_rows, missing_cols = np.where(np.isnan(data_clean))
data_clean[missing_rows, missing_cols] = column_means[missing_cols]

print("\nColumn means:")
print(np.round(column_means, 2))

print("\nCleaned data:")
print(data_clean)

print("Number of missing values after cleaning:", np.sum(np.isnan(data_clean)))

# 3. Separate X and y
X = data_clean[:, 0:3]
y = data_clean[:, 3]

print("\nX shape:", X.shape)
print("y shape:", y.shape)

# 4. Shuffle data
rng = np.random.default_rng(42)

indices = rng.permutation(len(data_clean))

X_shuffled = X[indices]
y_shuffled = y[indices]

# 5. Train-test split
train_size = int(0.8 * len(X_shuffled))

X_train = X_shuffled[:train_size]
y_train = y_shuffled[:train_size]

X_test = X_shuffled[train_size:]
y_test = y_shuffled[train_size:]

print("\nX_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

# 6. Feature scaling
train_mean = np.mean(X_train, axis=0)
train_std = np.std(X_train, axis=0)

X_train_scaled = (X_train - train_mean) / train_std
X_test_scaled = (X_test - train_mean) / train_std

print("\nTraining mean:", np.round(train_mean, 2))
print("Training std:", np.round(train_std, 2))

print("\nScaled X_train:")
print(np.round(X_train_scaled, 2))

print("\nScaled X_test:")
print(np.round(X_test_scaled, 2))

# 7. Manual linear prediction
weights = np.array([100, -5, 20])
intercept = 2500

y_pred = intercept + np.dot(X_test, weights)

print("\nActual y_test:")
print(y_test)

print("Predicted y_pred:")
print(np.round(y_pred, 2))

# 8. Error calculation
error = y_test - y_pred
absolute_error = np.abs(error)
squared_error = error ** 2

print("\nError:")
print(np.round(error, 2))

print("Absolute error:")
print(np.round(absolute_error, 2))

print("Squared error:")
print(np.round(squared_error, 2))

# 9. Metrics
mae = np.mean(absolute_error)
mse = np.mean(squared_error)
rmse = np.sqrt(mse)
mape = np.mean(np.abs(error / y_test)) * 100

print("\nMAE:", np.round(mae, 2))
print("MSE:", np.round(mse, 2))
print("RMSE:", np.round(rmse, 2))
print("MAPE:", np.round(mape, 2), "%")

# 10. Bad prediction detection
bad_prediction_mask = absolute_error > 200

bad_X = X_test[bad_prediction_mask]
bad_actual = y_test[bad_prediction_mask]
bad_predicted = y_pred[bad_prediction_mask]
bad_error = absolute_error[bad_prediction_mask]

print("\nBad prediction samples:")
print(bad_X)

print("Bad actual values:")
print(bad_actual)

print("Bad predicted values:")
print(np.round(bad_predicted, 2))

print("Bad absolute errors:")
print(np.round(bad_error, 2))

# 11. Final model status
print("\nFinal model status:")

if mape <= 5:
    print("Excellent model")
elif mape <= 10:
    print("Acceptable model")
else:
    print("Model needs improvement")
```
