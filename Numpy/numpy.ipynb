```python
actual = np.array([4200, 4300, 3900, 4100])
predicted = np.array([4150, 4400, 3850, 4000])

error = actual - predicted

print(error)
```

```python
absolute_error = np.abs(error)

print(absolute_error)
```

```python
mae = np.mean(absolute_error)

print(mae)
```
