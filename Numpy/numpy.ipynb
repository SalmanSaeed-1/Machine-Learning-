```python

data = np.array([
    [10, 30, 4200],
    [12, 35, 4300],
    [9, 40, 3900],
    [11, 32, 4100],
    [13, 29, 4500]
])

X = data[:, 0:2]
y = data[:, 2]

pressure = X[:, 0]

predicted = 3000 + 100 * pressure

error = y - predicted
absolute_error = np.abs(error)

bad_rows = data[absolute_error > 100]

print("Actual production:")
print(y)

print("Predicted production:")
print(predicted)

print("Absolute error:")
print(absolute_error)

print("Rows with error greater than 100:")
print(bad_rows)
```

```python
data = np.array([
    [10, 30, 4200],
    [12, 35, 4300],
    [9, 40, 3900],
    [11, 32, 4100],
    [13, 29, 4500]
])

X = data[:, 0:2]
y = data[:, 2]

pressure = X[:, 0]

predicted = 3000 + 100 * pressure

error = y - predicted
absolute_error = np.abs(error)

bad_prediction_rows = data[absolute_error > 100]

labels = np.where(absolute_error <= 100, "Good", "Bad")

print("Features X:")
print(X)

print("Target y:")
print(y)

print("Predicted:")
print(predicted)

print("Error:")
print(error)

print("Absolute Error:")
print(absolute_error)

print("Prediction Labels:")
print(labels)

print("Rows with bad predictions:")
print(bad_prediction_rows)
```

```python
actual = np.array([4200, 4300, 3900, 4100, 4500])
predicted = np.array([4000, 4200, 3900, 4100, 4300])

error = actual - predicted
absolute_error = np.abs(error)
squared_error = error ** 2

mae = np.mean(absolute_error)
mse = np.mean(squared_error)
rmse = np.sqrt(mse)
mape = np.mean(np.abs(error / actual)) * 100

ss_res = np.sum((actual - predicted) ** 2)
ss_total = np.sum((actual - np.mean(actual)) ** 2)
r2 = 1 - (ss_res / ss_total)

print("Actual:", actual)
print("Predicted:", predicted)
print("Error:", error)
print("Absolute Error:", absolute_error)
print("Squared Error:", squared_error)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", np.round(rmse, 2))
print("MAPE:", np.round(mape, 2), "%")
print("R²:", np.round(r2, 3))
```

```python
import numpy as np

actual = np.array([4200, 4300, 3900, 4100, 4500])
predicted = np.array([4000, 4200, 3900, 4100, 4300])

error = actual - predicted
absolute_error = np.abs(error)
squared_error = error ** 2

mae = np.mean(absolute_error)
mse = np.mean(squared_error)
rmse = np.sqrt(mse)
mape = np.mean(np.abs(error / actual)) * 100

ss_res = np.sum((actual - predicted) ** 2)
ss_total = np.sum((actual - np.mean(actual)) ** 2)
r2 = 1 - (ss_res / ss_total)

print("Error:", error)
print("Absolute Error:", absolute_error)
print("Squared Error:", squared_error)
print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", np.round(rmse, 2))
print("MAPE:", np.round(mape, 2), "%")
print("R²:", np.round(r2, 3))

if mape <= 5:
    print("Excellent model")
elif mape <= 10:
    print("Acceptable model")
else:
    print("Poor model")
```
