```python
def compute_cost(x , y , w , b):

    m = x.shape[0]
    
    cost = 0.0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost += (f_wb - y[i]) ** 2
    
    cost = cost/(2 * m)
    
    return cost

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

w = 200
b = 100

cost = compute_cost(x_train, y_train, w, b)

print(cost)
```

```python
def compute_gradient(x , y , w , b):
    dj_dw = 0.0
    dj_db = 0.0
    
    m = x.shape[0]
    
    for i in range(m):
        f_wb = w * x[i] + b
        error = (f_wb - y[i])
        
        dj_dw += (error * x[i])
        dj_db += error
        
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    
    return dj_dw , dj_db 

x_train = np.array([1.0, 2.0])

y_train = np.array([300.0, 500.0])

w = 200

b = 100

dj_dw, dj_db = compute_gradient(
    x_train,
    y_train,
    w,
    b
)

print(dj_dw)
print(dj_db)
```

```python
def gradient_descent(x , y, w_in , b_in , alpha , num_iters):
    w = w_in
    b = b_in
    j_history = []
    
    for i in range(num_iters):
        dj , dw = compute_gradient(x , y , w , b)
        
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        cost = compute_cost(x , y , w , b)
        j_history.append(cost)
        
        if i%100 == 0:
            print(f"Iterration{i:4}: Cost {cost:.2f}")
            
    return w , b , j_history

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])

w = 0
b = 0

alpha = 0.1
iterations = 1000

w , b , j= gradient_descent(
    x_train,
    y_train,
    w,
    b,
    alpha,
    iterations
)

print("\nFinal Parameters")
print("w =", w)
print("b =", b)

```

```python
def compute_cost(x , y , w, b):
    m = x.shape[0]
    f_wb = w * x + b
    error = f_wb - y
    squared_error = error ** 2
    cost = np.sum(squared_error)
    cost = cost / (2 * m)
    
    return cost
```

```python
def compute_gradient(x , y , w , b):
    m = x.shape[0]
    prediction = w * x + b
    error = prediction - y
    
    dj_dw_i = error * x
    dj_dw = np.sum(dj_dw_i)
    dj_db = np.sum(error)
    
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    
    return dj_dw , dj_db
    
    
```
