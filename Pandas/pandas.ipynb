```python
import numpy as np

data = {
    "Name": ["Ali", "Sara", "Omar", "Maha", "Ali", "Khalid", "Noor"],
    "Age": [25, 30, np.nan, 28, 25, 40, 120],
    "Gender": ["Male", "Female", "male", "Female", "Male", np.nan, "Female"],
    "Salary": [500, 700, 650, np.nan, 500, 900, 10000],
    "Department": ["IT", "Finance", "it", "HR", "IT", "Finance", "HR"],
    "Purchased": ["Yes", "No", "Yes", "No", "Yes", "Yes", "No"]
}

df = pd.DataFrame(data)
df
```

```python
df.isna().sum()

```

```python
df.isna().sum() * 100

```

```python
df_drop = df.dropna()
```

```python
df_drop
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```

```python
df
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
df
```

```python
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df["Gender"].unique()

df["Gender"] = df["Gender"].str.title()

```

```python
df["Gender"].unique()
```

```python
df
```

```python
df["Department"] = df["Department"].str.upper()

df["Department"].unique()
```

```python
df["Salary"] = pd.to_numeric(df["Salary"])
```
