```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x , y)
plt.show()
```

```python
fig , ax = plt.subplots()

ax.plot(x , y)

plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]
training_loss = [0.9, 0.7, 0.55, 0.42, 0.35]

fig , ax = plt.subplots()

ax.plot(epochs , training_loss , label= "Training loss")

ax.set_title("Training loss")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()

plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]

train_loss = [0.9, 0.7, 0.55, 0.42, 0.35]
val_loss = [1.0, 0.82, 0.70, 0.68, 0.72]

fig, ax = plt.subplots()

ax.plot(epochs, train_loss, label="Training Loss")
ax.plot(epochs, val_loss, label="Validation Loss")

ax.set_title("Training vs Validation Loss")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()

plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]
accuracy = [0.65, 0.72, 0.78, 0.83, 0.86]

fig , ax = plt.subplots(figsize=(8 , 5))

ax.plot(epochs , accuracy)

ax.set_title("Model accuracy over epoches")
ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")

plt.show()
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7]
exam_score = [45, 50, 55, 65, 70, 78, 85]

fig , ax = plt.subplots()

ax.scatter(study_hours , exam_score)

ax.set_title("Study hours vs Exam scores")

ax.set_xlabel("Study hours")
ax.set_ylabel("Exam scores")

plt.show()
```

```python
ages = [18, 19, 20, 20, 21, 22, 22, 23, 25, 30, 35, 40]

fig , ax = plt.subplots()

plt.hist(ages , bins=5)

ax.set_title("Age distribution")

ax.set_xlabel("Age")

ax.set_ylabel("Frequesncy")

plt.show()
```

```python
models = ["Linear Regression", "Random Forest", "XGBoost"]
accuracy = [0.78, 0.86, 0.90]

fig, ax = plt.subplots()

ax.bar(models, accuracy)

ax.set_title("Model Accuracy Comparison")
ax.set_xlabel("Model")
ax.set_ylabel("Accuracy")

plt.show()
```

```python

epochs = [1, 2, 3, 4, 5, 6]

train_accuracy = [0.60, 0.68, 0.75, 0.80, 0.84, 0.87]
val_accuracy = [0.58, 0.65, 0.72, 0.76, 0.77, 0.76]

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(epochs, train_accuracy, label="Training Accuracy")
ax.plot(epochs, val_accuracy, label="Validation Accuracy")

ax.set_title("Training vs Validation Accuracy")
ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")
ax.legend()

plt.show()
```
