```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x , y)
plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]
training_loss = [0.9, 0.7, 0.55, 0.42, 0.35]

fig , ax = plt.subplots()

ax.plot(epochs , training_loss , label= "Training loss")

ax.set_title("Training loss")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()

plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]

train_loss = [0.9, 0.7, 0.55, 0.42, 0.35]
val_loss = [1.0, 0.82, 0.70, 0.68, 0.72]

fig, ax = plt.subplots()

ax.plot(epochs, train_loss, label="Training Loss")
ax.plot(epochs, val_loss, label="Validation Loss")

ax.set_title("Training vs Validation Loss")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.legend()

plt.show()
```
